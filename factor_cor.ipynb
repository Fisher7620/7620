```python
import pandas as pd
import os
import numpy as np

#因子相关性计算

path = './feature/factor_final2/'
all_files = sorted(os.listdir(path))
array_2d = [[0 for i in range(4)] for j in range(4)]
n = 0

for filename in all_files:
    filename = path + filename
    df = pd.read_csv(filename, index_col=None, header=0)
    df = df.fillna(0)
    df = df.drop(df.columns[0], axis=1)
    corr = df.corr()
    if corr.isnull().any().any():
        continue
    m = np.array(corr.values)
    array_2d = array_2d + m
    n = n+1

array_2d = array_2d / n
df = pd.DataFrame(array_2d)
df[df==1] = 0
df.describe()
```

```python
import pandas as pd
import os
import numpy as np

class FactorCorrelationCalculator:
    def __init__(self, path):
        self.path = path
        self.all_files = sorted(os.listdir(path))
        self.array_2d = np.zeros((4, 4))
        self.n = 0

    def calculate_correlation(self):
        for filename in self.all_files:
            filename = os.path.join(self.path, filename)
            df = pd.read_csv(filename, index_col=None, header=0)
            df = df.fillna(0)
            df = df.drop(df.columns[0], axis=1)
            corr = df.corr()
            if corr.isnull().any().any():
                continue
            m = np.array(corr.values)
            self.array_2d += m
            self.n += 1

        self.array_2d /= self.n
        self.array_2d[self.array_2d == 1] = 0
        self.correlation_df = pd.DataFrame(self.array_2d)

    def describe_correlation(self):
        return self.correlation_df.describe()

path = './feature/factor_final2/'
calculator = FactorCorrelationCalculator(path)
calculator.calculate_correlation()
correlation_stats = calculator.describe_correlation()
print(correlation_stats)
```
